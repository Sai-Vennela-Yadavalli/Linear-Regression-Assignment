```python
# Pandas is the Library used to deal with the dataframes

import pandas as pd
```

```python
# Here we are assigning the whole data to the variable "Dataset" and by using Pandas we are loading the dataset.

Dataset = pd.read_csv("Salary_Data.csv")
```

```python
# Now we are just seeing what are all present in the Dataset
Dataset
```

```python
# Step 2 -> Data Preprocessing will be explained later
```

```python
# Step 3 -> Input Output split
# Here we are splitting and assigning the Input and Output to the variables "Independent"(Input) and "Dependent (Output) respectively

Independent = Dataset[["YearsExperience"]]
Dependent = Dataset[["Salary"]]
```

```python
# Step 4 -> Train Test split
# Here we are splitting Train and Test

from sklearn.model_selection import train_test_split

# We need to assign the Train and Test data to the Variables

x_train, x_test, y_train, y_test = train_test_split(Independent, Dependent, test_size = 0.3, random_state = 0)
```

```python
x_train, x_test, y_train, y_test
```

```python
# Here we selected the LinearRegression Algorithm, so we need to import that class
from sklearn.linear_model import LinearRegression

# We are assigning the whole class of Linear Regression to the variable "Regressor"
Regressor = LinearRegression()

# As we know, we have the formula here to calculate w and c, so we need to assign the values to the formula to calculate them
Regressor.fit(x_train, y_train)
```

```python
# If we want to know the w and c value, then we need to follow the below

Weight = Regressor.coef_
Weight
```

```python
Bias = Regressor.intercept_
Bias
```

```python
# Now we need to Predict

y_pred = Regressor.predict(x_test)

```

```python
# Now we need to evaluate

from sklearn.metrics import r2_score

r_score = r2_score(y_test, y_pred)
```

```python
r_score
```

```python
# Now we need to save our model
# As of now entire model is saved in Regressor

import pickle

filename = "Best Model.sav"
```

```python
# Now we need to load/dump our model into the file we just created

pickle.dump(Regressor, open(filename, "wb"))
```

```python
# Now lets load and see whether the model has been loaded correctly

Loaded_model = pickle.load(open(filename, "rb"))
```

```python
# Now we need to get the inputs

Result = Loaded_model.predict([[13]])
```
